First for all float variables. Find classwise distribution of NA's.

```python
float_cols = [col for col in train_df.columns if train_df[col].dtype == 'float64']
int_cols = [col for col in train_df.columns if train_df[col].dtype == 'int64']
object_cols = [col for col in train_df.columns if train_df[col].dtype == 'O']
assert len(float_cols + int_cols + object_cols) == len(train_df.columns)
```

```python
na_float_df = pd.DataFrame(index = float_cols, columns = ['Class 0', 'Class 1'])
for i, col in enumerate(float_cols):
    
    perc_1 = train_df[train_df['convert_30'] == 1][col].isna().sum()/train_df[train_df['convert_30'] == 1][col].shape[0]
    perc_0 = train_df[train_df['convert_30'] == 0][col].isna().sum()/train_df[train_df['convert_30'] == 0][col].shape[0]
    
    na_float_df.iloc[i] = [perc_0, perc_1]
```

```python
na_float_df
```

```python
na_int_df = pd.DataFrame(index = int_cols, columns = ['Class 0', 'Class 1'])
for i, col in enumerate(int_cols):
    
    perc_1 = train_df[train_df['convert_30'] == 1][col].isna().sum()/train_df[train_df['convert_30'] == 1][col].shape[0]
    perc_0 = train_df[train_df['convert_30'] == 0][col].isna().sum()/train_df[train_df['convert_30'] == 0][col].shape[0]
    
    na_int_df.iloc[i] = [perc_0, perc_1]
```

```python
# No NA's here
na_int_df
```

```python
na_obj_df = pd.DataFrame(index = object_cols, columns = ['Class 0', 'Class 1'])
for i, col in enumerate(object_cols):
    
    perc_1 = train_df[train_df['convert_30'] == 1][col].isna().sum()/train_df[train_df['convert_30'] == 1][col].shape[0]
    perc_0 = train_df[train_df['convert_30'] == 0][col].isna().sum()/train_df[train_df['convert_30'] == 0][col].shape[0]
    
    na_obj_df.iloc[i] = [perc_0, perc_1]
```

```python
# No NA's here too
na_obj_df
```

Loads of missing values here for float/continuous features.
Imputation approach: There's 2 classes here: customers who converted vs those who didn't. These features are intuitively correlated to the outcome so we can subset the training data for those who didn't convert and impute all NA's to 0. For those who did convert we can impute median values since there was most likely some past activity. The only way to be sure is to try different imputation methods and check validation scores but for now, this seems like the safest assumption to make instead of imputing all NA's with median values.

```python
# Subset by class
train_df_0 = train_df[train_df['convert_30'] == 0]
train_df_1 = train_df[train_df['convert_30'] == 1]
```

```python
# Fill NA's with 0's
train_df_0 = train_df_0.fillna(0)

# Fill NA's with medians
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'median', axis = 0)
values_1 = train_df_1[float_cols].values
imputed_1 = imputer.fit_transform(values_1)

# Write these back into train_df_1
for i, col in enumerate(float_cols):
    
    train_df_1[col] = imputed_1[:,i]
```

```python
train_df_nonas = train_df_0.append(train_df_1)
```

```python
train_df_nonas_ohe = pd.get_dummies(train_df_nonas)
```

### Oversampling the minority class.
We'll split into train and test first, then Oversample. OVersampling before splitting allows similar samples from train to bleed into test and bias results.

```python
from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=0)
X = train_df_nonas_ohe.drop(['cuid', 'convert_30', 'revenue_30'], axis=1).values
y = train_df_nonas_ohe['convert_30'].values
sss.get_n_splits(X, y)
for train_index, test_index in sss.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=12, ratio = 0.8)
x_train_res, y_train_res = sm.fit_sample(X_train, y_train)
```

```python
np.unique(y_train_res, return_counts=True)
```

### Finding highly correlated features and leaving them out

```python
x_train_res_df = pd.DataFrame(data=np.hstack((x_train_res,y_train_res.reshape(len(y_train_res),1))), columns=list(train_df_nonas_ohe.columns[3:]) + ['convert_30'])
corr = x_train_res_df.corr()
```

```python
#Correlation with output variable
cor_target = abs(corr['convert_30'])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.2]
relevant_features
```

```python
temp_features = [i[0] for i in relevant_features.items()]
```

```python
Feature_SelectionSet1 = x_train_res_df[temp_features]
```

```python
fig = plt.figure(figsize=(20,20))
corr = Feature_SelectionSet1.corr()
sns.heatmap(corr, annot=True, cmap=plt.cm.Reds)
plt.show()
fig.savefig("Correlation matrix for Set 1")
```

One of the assumptions of machine learning models is that the independent variables need to be uncorrelated with each other. If these variables are correlated with each other, then we need to keep only one of them and drop the rest. So let us check the correlation of selected features with each other. This can be done either by visually checking it from the above correlation matrix 

```python
corr_matrix = corr.abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]
```

```python
to_drop
```

```python
Feature_SelectionSet1=Feature_SelectionSet1.drop(Feature_SelectionSet1[to_drop], axis=1)
```
